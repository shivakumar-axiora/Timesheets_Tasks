```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="darkgrid")

df = pd.read_csv("StudentsPerformance.csv")
```

```python
print("Top5 Rows:\n",df.head())
print("\nInfo:\n",df.info())
print("\nDescribe:\n",df.describe())


```

**1. Histogram — Math Score Distribution**

Question: Are most students performing average, low, or high in math?

```python
plt.figure()
sns.histplot(df['math score'], kde=True)
plt.title("Distribution of Math Scores")
plt.show()
```

Most students cluster around mid-range scores, with fewer high achievers.

**2. Scatter Plot — Reading vs Writing**

Question: Do reading skills influence writing performance?

```python
plt.figure()
sns.scatterplot(x='reading score', y='writing score', data=df)
plt.title("Reading vs Writing Scores")
plt.show()
```

Strong positive relationship → better readers are better writers.

**3. Bar Plot — Average Scores by Gender**

Question: Is there a performance gap between genders?

```python
df.groupby('gender')[['math score','reading score','writing score']].mean().plot(kind='bar')
plt.title("Average Scores by Gender")
plt.xticks(rotation=0)
plt.show()
```

Females often outperform in reading/writing; math is closer.

**4. Box Plot — Math Score by Lunch Type**

Question: Does economic background (lunch type) impact performance?

```python
plt.figure()
sns.boxplot(x='lunch', y='math score', data=df)
plt.title("Math Score by Lunch Type")
plt.show()
```

Students with standard lunch perform better → economic factor matters.

**5. Violin Plot — Scores by Test Prep Course**

Question: Does test preparation improve scores?

```python
plt.figure()
sns.violinplot(x='test preparation course', y='math score', data=df)
plt.title("Impact of Test Prep on Math Score")
plt.show()
```

Students who completed prep show higher and more consistent scores.

**6. Heatmap — Correlation**

Question: Which subjects are strongly related?

```python
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Between Scores")
plt.show()
```

Math, reading, and writing are strongly correlated.

**7. Pair Plot — Overall Relationships**

Question: What patterns exist across all scores?

```python
sns.pairplot(df[['math score','reading score','writing score']])
plt.show()
```

Clear linear relationships between all three subjects.

**8. Line Plot (Simulated Trend)**

Question: How do average scores vary across parental education levels?

```python
avg_scores = df.groupby('parental level of education')['math score'].mean()

plt.figure()
avg_scores.plot(kind='line', marker='o')
plt.title("Math Score by Parental Education")
plt.xticks(rotation=45)
plt.show()
```

Higher parental education often correlates with better student performance.

**9. Count Plot — Test Prep Participation**

Question: How many students take preparation courses?

```python
sns.countplot(x='test preparation course', data=df)
plt.title("Test Preparation Participation")
plt.show()
```

Majority may not take prep → opportunity for improvement programs.

**Final Insights :**

The analysis reveals that student performance is not driven by a single factor, but a combination of academic ability, socio-economic conditions, and preparation efforts.
